# Evolution of actors over time

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_movies = pd.read_pickle("../data/generated/preprocessed/movies.pkl")
df_chars = pd.read_pickle("../data/generated/preprocessed/characters.pkl")
df_actors = pd.read_pickle("../data/generated/preprocessed/actors.pkl")
```

## General actor data

```python
df_actor_mins = df_chars[["fb_actor_id", "actor_age", "movie_release_year"]].groupby("fb_actor_id").min()
df_actor_mins.rename(columns = {'actor_age':'debut_age', 'movie_release_year':'debut_year'}, inplace=True)
df_actors = pd.merge(df_actors, df_actor_mins, on="fb_actor_id", how="left")
```

```python
df_actor_maxs = df_chars[["fb_actor_id", "actor_age", "movie_release_year"]].groupby("fb_actor_id").max()
df_actor_maxs.rename(columns = {'actor_age':'last_movie_age', 'movie_release_year':'last_movie_year'}, inplace=True)
df_actors = pd.merge(df_actors, df_actor_maxs, on="fb_actor_id", how="left")
```

```python
df_actors_in_movies = df_chars.drop_duplicates(subset=["fb_actor_id", "fb_movie_id"])
df_actor_movie_count = df_chars[["fb_actor_id"]].groupby("fb_actor_id").size().astype("UInt32")
df_actor_movie_count.name = "movies_count"
df_actors = pd.merge(df_actors, df_actor_movie_count, on="fb_actor_id", how="left")
```

```python
df_actors["movies_per_decade"] = (df_actors.movies_count * 10) / (1 + df_actors.last_movie_year - df_actors.debut_year)
```

```python
mpd_max = 20
mpd_bins = 10
mpd_bin_size = (mpd_max // mpd_bins)
mpd_actor_counts = []
for ix in range(mpd_bins):
    mpd_actor_counts.append(df_actors.loc[(df_actors.movies_per_decade >= mpd_bin_size*ix) & (df_actors.movies_per_decade < mpd_bin_size*(ix+1))].size)
```

```python
for i,v in enumerate(mpd_actor_counts):
    print("[%d, %d[: %d"%(mpd_bin_size*i, mpd_bin_size*(i+1), v))
```

```python
without_outlier = df_actors.movies_per_decade.dropna().loc[df_actors.movies_per_decade < 30]
fig, ax = plt.subplots()
ax.set_xlabel("Number of movies played in per decade")
ax.set_ylabel("Number of actors")
ax.set_yscale("log")
ax.set_xticks(range(0, 30, 2))
ax.set_title("Movies per decades")
hist = ax.hist(without_outlier, bins=30)
```
