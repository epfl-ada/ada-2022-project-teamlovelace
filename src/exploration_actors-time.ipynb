# Evolution of actors over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_movies = pd.read_pickle("../data/generated/preprocessed/movies.pkl")
df_roles = pd.read_pickle("../data/generated/preprocessed/roles.pkl")
df_actors = pd.read_pickle("../data/generated/preprocessed/actors.pkl")
```

```python
df_all = pd.merge(df_roles, df_movies, left_on="fb_movie_id", right_on="fb_id")
df_all
```

## General actor data

```python
df_actor_mins = df_all[["fb_actor_id", "actor_age", "movie_release_year"]].groupby("fb_actor_id").min()
df_actor_mins.rename(columns = {'actor_age':'debut_age', 'movie_release_year':'debut_year'}, inplace=True)
df_actors = pd.merge(df_actors, df_actor_mins, on="fb_actor_id", how="left")
```

```python
df_actor_maxs = df_all[["fb_actor_id", "actor_age", "movie_release_year"]].groupby("fb_actor_id").max()
df_actor_maxs.rename(columns = {'actor_age':'last_movie_age', 'movie_release_year':'last_movie_year'}, inplace=True)
df_actors = pd.merge(df_actors, df_actor_maxs, on="fb_actor_id", how="left")
```

```python
df_actors_in_movies = df_roles.drop_duplicates(subset=["fb_actor_id", "fb_movie_id"])
df_actor_movie_count = df_actors_in_movies[["fb_actor_id"]].groupby("fb_actor_id").size().astype("UInt32")
df_actor_movie_count.name = "movies_count"
df_actors = pd.merge(df_actors, df_actor_movie_count, on="fb_actor_id", how="left")
```

```python
df_actor_med = df_all[["fb_actor_id", "tomato_score", "movie_revenue"]].groupby("fb_actor_id").median()
df_actor_med.rename(columns = {'tomato_score':'median_tomato_score', 'movie_revenue':'median_movie_revenue'}, inplace=True)
df_actors = pd.merge(df_actors, df_actor_med, on="fb_actor_id", how="left")
```

```python
df_actors["movies_per_decade"] = (df_actors.movies_count * 10) / (1 + df_actors.last_movie_year - df_actors.debut_year)
```

```python
df_all = pd.merge(df_all, df_actors, on="fb_actor_id")
```

```python
df_actors.movies_per_decade.describe()
```

```python
mpd_max = 20
mpd_bins = 10
mpd_bin_size = (mpd_max // mpd_bins)
mpd_actor_counts = []
for ix in range(mpd_bins):
    mpd_actor_counts.append(df_actors.loc[(df_actors.movies_per_decade >= mpd_bin_size*ix) & (df_actors.movies_per_decade < mpd_bin_size*(ix+1))].size)
```

```python
for i,v in enumerate(mpd_actor_counts):
    print("[%d, %d[: %d"%(mpd_bin_size*i, mpd_bin_size*(i+1), v))
```

```python
without_outlier = df_actors.movies_per_decade.dropna().loc[df_actors.movies_per_decade < 30]
fig, ax = plt.subplots()
ax.set_xlabel("Number of movies played in per decade")
ax.set_ylabel("Number of actors")
ax.set_yscale("log")
ax.set_xticks(range(0, 30, 2))
ax.set_title("Movies per decades")
hist = ax.hist(without_outlier, bins=30)
```

## See how the movie ratings of actors evolve over their careers

```python
df_big_actors = df_all.loc[df_all.movies_count >= 10]
```

```python
df_role_tomato = df_big_actors[["tomato_score", "movie_release_year", "debut_year", "median_tomato_score"]].dropna()
df_role_tomato["time_in_career"] = df_role_tomato.movie_release_year - df_role_tomato.debut_year
df_role_tomato["relative_score"] = df_role_tomato.tomato_score - df_role_tomato.median_tomato_score
df_role_tomato[["time_in_career", "relative_score"]]

sns.jointplot(x=df_role_tomato.time_in_career, y=df_role_tomato.relative_score, kind="hex")
```

```python
df_role_money = df_big_actors[["movie_revenue", "movie_release_year", "debut_year", "median_movie_revenue"]].dropna()
df_role_money["time_in_career"] = df_role_money.movie_release_year - df_role_money.debut_year
df_role_money["relative_money"] = df_role_money.movie_revenue / df_role_money.median_movie_revenue
df_role_money = df_role_money[["time_in_career", "relative_money"]]
df_role_money = df_role_money.loc[df_role_money.relative_money < 10]
df_role_money
```

```python
g = sns.jointplot(x=df_role_money.time_in_career, y=df_role_money.relative_money, kind="hex")
```
