```python
import pandas as pd
import json
```

```python
def flatten_json_dict(js_dict):
    return ":".join(json.loads(js_dict).values())
```

```python
plot_summaries = pd.read_table('dataset/plot_summaries.txt', header=None, names=["wiki_id", "summary"], index_col=0)

#names_clusters maps character_metadata entries to character names. (probably useless)
names_clusters = pd.read_table('dataset/name.clusters.txt', header=None, names=["character", "fb_id"])
names_clusters = names_clusters.drop_duplicates()
names_clusters.index = names_clusters.fb_id
names_clusters = names_clusters[["character"]]

tv_tropes_clusters = pd.read_table('dataset/tvtropes.clusters.txt', header=None)
tv_tropes_clusters = pd.concat(objs=[tv_tropes_clusters[[0]], pd.json_normalize(tv_tropes_clusters[1].apply(json.loads))], axis=1)
tv_tropes_clusters.columns = ["trope", "character", "movie", "fb_char_id", "actor"]

movies_metadata = pd.read_table('dataset/movie.metadata.tsv', header=None, names=["wiki_id", "fb_id", "name", "release_date", "revenue", "runtime", "language", "country", "genre"])
movies_metadata.language = movies_metadata.language.apply(flatten_json_dict)
movies_metadata.country = movies_metadata.country.apply(flatten_json_dict)
movies_metadata.genre = movies_metadata.genre.apply(flatten_json_dict)

# I don't know what the column "something" means
character_metadata = pd.read_table('dataset/character.metadata.tsv', header=None, names=["wiki_movie_id", "fb_movie_id", "movie_date", "character", "date_of_birth", "gender", "height", "ethnicity", "actor", "actor_age", "fb_id", "something", "something_else"], index_col="fb_id")
```
